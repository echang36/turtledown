```python
import nltk
from nltk.book import *
```

```python
#import stopwords into a variable
from nltk.corpus import stopwords
stopw=stopwords.words('english')
```

```python
state_union = nltk.corpus.state_union
```

```python
state_union.fileids()
```

```python
fileid ='1953-Eisenhower.txt'
speech = [w.lower() for w in set(state_union.words(fileid)) if w not in stopw]
fdist = FreqDist(speech)
fdist.items()
#fdist(25, cumulative=True)
```

```python
from nltk.corpus import inaugural
inaugural.fileids()
```

```python
fileid ='2005-Bush.txt'
speech = [w.lower() for w in set(inaugural.words(fileid)) if w not in stopw and w.isalpha()]
fdist = FreqDist(speech)
fdist.items()
```
